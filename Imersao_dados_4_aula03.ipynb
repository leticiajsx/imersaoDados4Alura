```python
dados_vendas.describe()
```

```python
outliers_metragem = dados_vendas['Metragem'] <= 450
outliers_quartos = (dados_vendas['Quartos'] >= 2) & (dados_vendas['Quartos'] <= 5)
outliers_banheiros = dados_vendas['Banheiros'] <= 5
outliers_vagas = dados_vendas['Vagas'] <= 5
outliers_valor_anuncio = dados_vendas['Valor_anuncio_float'] <= 3500000
outliers_valor_m2 = (dados_vendas['Valor_m2'] >= 1000) & (dados_vendas['Valor_m2'] <= 9453.78)

dados_vendas[outliers_metragem & outliers_quartos & outliers_banheiros & outliers_vagas & outliers_valor_anuncio & outliers_valor_m2]
```

```python
dados_agrupados_metragem = dados.groupby(['Metragem']).mean()
dados_agrupados_metragem = dados_agrupados_metragem.reset_index()
dados_agrupados_metragem = dados_agrupados_metragem.nlargest(n=10, columns=['Metragem'])
dados_agrupados_metragem
```

```python
sns.catplot(x="Metragem", y='Quartos', kind="bar", palette="ch:.25", data=dados_agrupados_metragem, height=7, aspect=2)
```
