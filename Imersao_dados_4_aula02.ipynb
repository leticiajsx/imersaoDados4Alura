# Desafios

1 - Criar uma escala de valores em milhões

```python
dados_vendas['Valor_milhoes'] = dados_vendas['Valor_anuncio_float']/1000000
dados_vendas.describe()
```

2 - Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

```python
sns.set()
plt.figure(figsize=(15,8))
sns.set_theme(style="darkgrid", palette="pastel")
ax = sns.histplot(data = dados_vendas, x='Valor_milhoes', bins=100, kde=True, color='#c88f1c')
ax.set_title('Histograma Valor Imóvel', fontsize='16')
ax.set_xlabel('Valor Imóveis em Milhões', fontsize='14')
ax.set_ylabel('Quantidade de Imóveis', fontsize='14')
plt.xlim(0, 14.5)
plt.xticks(list(range(0,15)))
plt.show()
```

3 - Preço do metro quadrado por bairro e plotar em um gráfico ideal

```python
dados_vendas['Valor_m2'] = dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem']
dados_bairro_valor_m2 = dados_vendas.groupby(['Bairro']).mean()
dados_bairro_valor_m2
```

```python
dez_maiores_valores_m2 = dados_bairro_valor_m2.nlargest(n=10, columns=['Valor_m2'])
dez_maiores_valores_m2['Valor_m2_milhares'] = dez_maiores_valores_m2['Valor_m2']/1000
dez_maiores_valores_m2 = dez_maiores_valores_m2.reset_index()
dez_maiores_valores_m2
```

```python
sns.set()
plt.figure(figsize=(20,8))
sns.set_theme(style="whitegrid")

ax = sns.scatterplot(data=dez_maiores_valores_m2, x="Valor_m2_milhares", y="Bairro", hue="Bairro")
ax.set_title('Top 10 bairros em São Paulo com m2 mais caro', fontsize='16')
ax.set_xlabel('Valor m2 em milhares', fontsize='14')
ax.set_ylabel('Bairros', fontsize='14')
plt.show()
```

4 - Explorar as bibliotecas de visualizações e colocar as suas conclusão

```python
cem_menores_valores_m2 = dados_vendas.nsmallest(n=100, columns=['Valor_m2'])
cem_menores_valores_m2['Valor_m2_milhares'] = cem_menores_valores_m2['Valor_m2']/1000
cem_menores_valores_m2 = cem_menores_valores_m2.reset_index()
cem_menores_valores_m2
```

```python
import plotly.express as px

fig = px.scatter_3d(cem_menores_valores_m2, x=cem_menores_valores_m2['Quartos'], y=cem_menores_valores_m2['Vagas'], z=cem_menores_valores_m2['Valor_m2_milhares'],
              color='Bairro')
fig.show()
```

Após analisar esse gráfico concluí que o bairro Jardim Guedala tem o menor valor de m2 na cidade de São Paulo
